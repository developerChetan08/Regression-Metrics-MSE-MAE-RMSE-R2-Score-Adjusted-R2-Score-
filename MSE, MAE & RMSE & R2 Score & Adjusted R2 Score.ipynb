```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
df.shape
```

# Create A Scatter Plot

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

# Apply train_test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

# Apply LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

# Create A Scatter Plot & Best fit Line

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

# Calculate All Regression Metrics

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

# y_test.values

# 1) Calculate Mean Absolute Error(MAE):-

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

# 2) Calculate Mean Squared Error:-

```python
print("MSE",mean_squared_error(y_test,y_pred))
```

# 3) Calculate Root Mean Squared Error:-

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

# 4) Calculate R2 Score:-

```python
print("R2 Score",r2_score(y_test,y_pred))
```

# 5) Calculate Adjusted R2 Score:-

```python
r2 = r2_score(y_test,y_pred)
```

```python
X_test.shape
```

```python
##Apply FOrmula:-
Adjusted_R2_Score= 1 - ((1-r2)*(40-1)/(40-1-1))
print("Adjusted R2 Score",Adjusted_R2_Score)
```
